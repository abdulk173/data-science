```python
import pandas as pd
import numpy as np
```

```python
data = {
    "EmpId": [101, 102, "103", 104, 105, 106, 107, 108, 109, 110],
    "fullName": [
        "  Alice Johnson  ", "MICHEAL smith", "sarah lee", "David-Kim", None, "john brown",
        "Emma Wilson", "TOM  davis", "Jane Miller", "Chris taylor"
    ],
     "Dept": ["IT", "hr", "Mktng", "IT", "Customer Care", "marketing", "Finance", None, "HR", "marketing"],
    "Age": [29, "thirty-five", 26, 31, 42, 27, 33, 30, None, 28],
    "Salary($)": ["$5500.50", "4800", "4200.75", "NaN", "6100", "4,300", "5800", "5000", "4700.0", "4400"],
    "JoinDate": [
        "2021/04/12", "2020/11/01", "2022/01/18", "2019/9/25", "2018/07/05",
        "2021/10/14", "2019/12/01", "2020/03/22", "2021/05/10", "2022/02/05"
    ],
    "Performance": [87, "73", 90, 65, 82, 75, 88, None, "79%", "91"],
    "Remote?": ["Yes", "No", "Y", "N", "yes", "no", "Y", "Yes", None, "True"]
}
```

```python
df = pd.DataFrame(data)

df.to_csv("employee_data.csv" , index=False)

df = pd.read_csv("employee_data.csv")

df
```

```python
# all info about the data
df.info()
```

```python
#rename inconsistent column

df.columns
df.rename(columns = {
    "EmpId": "Emp_id",
    "fullName": "Full_name", 
    "Dept": "Department",
    "Age": "Age",
"Salary($)": "Salary",
"JoinDate": "Join_date",
"Performance": "Performance",
"Remote?": "Remote",
}, inplace=True)

df
```

```python
df["Full_name"].tolist()
```

```python
#remove beginning and end spaces 

df["Full_name"] = df["Full_name"].str.strip()

df["Full_name"].tolist()
```

```python
#reduce multiple inside space to only one

df["Full_name"] = df["Full_name"].str.replace(r"\s+" , " ", regex=True)

df["Full_name"].tolist()
```

```python
#replace - with " "

df["Full_name"] = df["Full_name"].str.replace("-"," ")

df["Full_name"].tolist()
```

```python
#convert text to first letter upper case other small letter 

df["Full_name"] = df["Full_name"].str.title()

df["Full_name"].tolist()
```

```python
df
```

```python
df["Department"].tolist()
```

```python
df["Department"] = df["Department"].replace({
    "hr" : "HR",
    "Mktng" : "Marketing",
    "marketing" : "Marketing",
})

df["Department"].tolist()
```

```python
df
```

```python
def clean_age(value):
    if value == "thirty-five":
       return 35
    elif isinstance(value ,(int , float)):
        return value
    elif isinstance(value, str) and value.isdigit():
        return int(value)
    else : 
          return np.nan
df["Age"] = df["Age"].apply(clean_age)

df["Age"]
```

```python
df
```

```python
df["Salary"].tolist()
```

```python
df["Salary"] = (
    df["Salary"]
    .astype(str)
    .replace(r"[$ , ]" , "" , regex=True)
    .replace("NaN" , np.nan)
    .astype(float)
)

df["Salary"].tolist()
```

```python
df
```

```python
df["Join_date"] = pd.to_datetime(df["Join_date"] , errors="coerce")
df["Join_date"]
```

```python
df
```

```python
df["Performance"] = (
    df["Performance"]
    .astype(str)
    .replace(r"[%]","", regex=True)
    .replace("NaN", np.nan)
    .astype(float)
)

df["Performance"]
```

```python
df
```

```python
df["Remote"] = (
    df["Remote"]
    .astype(str)
    .replace({
        "Y":"Yes",
        "N":"No",
        "yes":"Yes",
        "no":"NO",
        "True":"Yes",
    })
    .replace("NaN", np.nan)
)

df["Remote"]
```
