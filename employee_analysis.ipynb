```python
# stat cleaning inconsistent value

#rename inconsistent column

df.columns
df.rename(columns = {
    "EmpId": "Emp_id",
    "fullName": "Full_name", 
    "Dept": "Department",
    "Age": "age",
"Salary($)": "salary",
"JoinDate": "join_date",
"Performance": "performance",
"Remote?": "remote",
}, inplace=True)

df
```

```python
df["full_name"] = df["Full_name"].str.strip()
df["full_name"] = df["Full_name"].str.title()
df["full_name"] = df["Full_name"].str.replace(r"\s+", " ", regex=True)
df["full_name"] = df["Full_name"].str.replace(r"-" , " ")
df["full_name"].tolist()
```

```python
#clean inconsistent value

df["Department"] = df["Department"].replace({
    "hr":"HR",
    "Mktng": "Marketing",
    "marketing": "Marketing"
    
 }, inplace=True)
```

```python
#clean age

def clean_age(value):
    if value == "thirty-five":
        return 35
    elif isinstance(value, (int, float)):
       return value
    elif isinstance(value, str) and value.isdigit():
       return int(value)
    else : 
          return np.man

df["age"] = df["age"].apply(clean_age) 

df["age"]


        
```

```python
#clean inconsistent salary

df["salary"] = (
    df["salary"]
    .astype(str)
    .replace(r"[$ , ]", "" , regex=True)
    .replace("NAN", np.nan)
    .astype(float)
)

df["salary"].unique()
```

```python
df["join_date"] = pd.to_datetime(df["join_date"] , errors = "coerce")
df["join_date"]
```

```python
#clean inconsistent value in performance

df["performance"] = (
    df["performance"]
    .astype(str)
    .str.replace("%"," ", regex=False)
    .replace("None", np.nan)
    .astype(float)
    )

df["performance"]
```

```python
#clean inconsistent value on remote

df["remote"].replace({
    "Y": "Yes",
    "N": "No",
    "no": "No",
    "True": "Yes",
    "yes": "Yes",
    None: np.nan
}, inplace=True)

df
```

```python
#make all number both string and int all numerical

df["Emp_id"] = pd.to_numeric(df["Emp_id"], errors="coerce")
df["Emp_id"]
```

```python
#Handling missing valueby replacing nan with unknown 
df["Full_name"] = df["Full_name"].fillna("unknown")
df["Full_name"]
```
