```python
#work with daily steps analysis using numpy

np.random.seed(42) #reproducibility
days = np.arange(1,31)
steps = np.random.randint(3456, 12034 , size = 30)

data = pd.DataFrame({
    "days": days,
    "steps": steps
})

data
```

```python
#basic stat

mean_steps = np.mean(steps)
std_steps = np.std(steps)
print("average steps:" , mean_steps)
print("standard deviation:" , std_steps)
print("median steps:", np.median(steps))
print("minimum steps:" , np.min(steps))
print("maximum steps:" , np.max(steps))
```

```python
#goal analysis steps exceed 10,000

goal = np.where(steps > 10000)[0] +1
print(goal)
```

```python
#weekly analysis(split)
weeks = np.array_split(steps , 4)
print("weeks")
print(weeks)
print("week total")
week_total = [print(week.sum())for week in weeks]
print("\nweek average")
week_average = [print(week.mean())for week in weeks] 

```

```python
window = 7
moving_average = np.convolve(steps, np.ones(window)/window, mode = "valid")
print("rolling movint average", moving_average[:7])
```

```python
#anomaly detection

lower_bound = mean_steps - 2*std_steps
upper_bound = mean_steps + 2*std_steps

print(lower_bound)
print(upper_bound)

anomaly = np.where((steps < lower_bound) | (steps > upper_bound))[0] + 1

print("anomaly detection low/high:" , anomaly)
```
