```python
#remove of spaces around a word
df["full_name"] = df["full_name"].str.strip()
df["full_name"].tolist()
```

```python
#replacing - with space

df["full_name"] = df["full_name"].str.replace(r"-" , " ")
df["full_name"].tolist()
```

```python
#replace multiple space with single space

df["full_name"] = df["full_name"].str.replace(r"\s+", " ", regex=True)
df["full_name"].tolist()
```

```python
#replace inconsistent element in department column

df["department"]= df["department"].replace({
    "hr":"HR",
    "Mktng":"Marketing",
    "marketing":"Marketing",
    "NaN":np.nan
})
df["department"].tolist()
```

```python
#list ao all elemnt in age

df["age"]
```

```python
#clean age

def clean_age(value):
    if value == "thirty-five":
        return 35
    elif isinstance(value, (int, float)):
       return value
    elif isinstance(value, str) and value.isdigit():
       return int(value)
    else : 
          return np.man

df["age"] = df["age"].apply(clean_age) 

df["age"]
```

```python
#clean inconsistent element salary

df["salary"] = (
    df["salary"]
    .astype(str)
    .replace(r"[$ , ]", "" , regex=True)
    .replace("NAN", np.nan)
    .astype(float)
)

df["salary"].unique()
```

```python
df["join_date"] = pd.to_datetime(df["join_date"] , errors = "coerce")
df["join_date"]
```

```python
#clean inconsistent value in performance

df["performance"] = (
    df["performance"]
    .astype(str)
    .str.strip()
    .str.replace("%", "", regex=False)
    .replace(["None", "nan", "N/A", "--", ""], np.nan)
    .astype(float)
)
df["performance"]
```

```python
#clean inconsistent value on remote

df["remote"]=df["remote"].replace({
    "Y": "Yes",
    "N": "No",
    "no": "No",
    "True": "Yes",
    "yes": "Yes",
    "NaN":np.nan
})
df["remote"].tolist()
```

```python
#detect an outlier 

num = merge_df[["age", "salary", "performance" , "department_budget", "department_avg_salary"]]
Q1 = num.quantile(0.25)
Q3 = num.quantile(0.75)
IQR = Q3 - Q1

lower_bond = Q1 - 1.5*IQR
upper_bond = Q3 + 1.5*IQR
outliers = (num<lower_bond) | (num>upper_bond)

print(outliers.sum())
```
